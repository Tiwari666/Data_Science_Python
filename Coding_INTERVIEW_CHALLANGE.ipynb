# **args example**

```python
def get_rmse(*args):
    n = len(args)
    sum_serr = sum((a - p) ** 2 for a, p in args)
    msqerr = sum_serr / n
    rmsqerr = msqerr ** 0.5
    return rmsqerr

# Example usage:
result = get_rmse((2, 2.1), (3, 3.2), (4, 4.5), (5, 4.9)) # in place of tuple points, list can also be used.
print(result)
```

# **Palindrome:**

```python
def palindrome(n):
  str_num = str(n)
  reverse_str = str_num[::-1]
  reverse_num = float(reverse_str)
  if n == reverse_num:
    return str(reverse_num) + ' is a palindrome.'
  else:
    return str(reverse_num) + " is not a palindrome."
```

```python
palindrome(122)
```

# **STRING REVERSE:**

```python
def reverse_str(s):
  result = ''
  for i in range(len(s)-1, -1, -1): #i = 2,1,0 for RAM
     result += s[i]
  return result

# case
#reverse_str('RAM')
reverse_str("412")
```

# **NUMBER TO ROMAN:**

```python
def get_roman(n):
  roman_list = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
  number_list = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]

  result = ''
  for i in range(0, len(roman_list)):
    while n >= number_list[i]: # keep looping until number_list >=1
      n -= number_list[i]
      result += roman_list[i]

  return result
```

```python
get_roman(4656)
```

# **ALTERNATIVE:**

```python
class Roman: #
  rlist = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I'] # class variable
  nlist = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]  # class variable

  def __init__(self,n): # default constructor defines istance varibles only.
    self.n =n

  def get_num_to_roman(self):
    result = ''
    for i in range(0, len(Roman.rlist)):
      while self.n>=Roman.nlist[i]:
        result += Roman.rlist[i]
        self.n-=Roman.nlist[i]
    return result
```

# **CLASS METHOD, OBJECT METHOD AND INSTANCE METHOD:**

```python
# @classmethod: It's primarily used  to define a class method within a class.
# The method defined using @classmethod receives the class itself (conventionally named cls)
#  as the first argument rather than an instance of the class (self).
# It can be called on the class itself, and it operates on the class and its attributes, not on instances

# object = instance

# Types of method:
# A) CLASS METHOD: DENOTED BY @classmethod--ANNOTATION (TO FIX/SET THE TYPE OF METHOD).
# We need to call the class method using the "class name", not the object/instance. e.g.: --Class_name.method_name= MyClass.show_class_variable()

# B)INSTANCE METHOD:

# We need to call the INSTANCE METHOD using the object/instance. e.g.:  obj.method_name()

class MyClass:
    class_variable = 10

    def __init__(self, value):
        self.value = value

    @classmethod
    def show_class_variable(cls):
        print(cls.class_variable)

    @classmethod
    def change_class_variable(cls, new_value):
        cls.class_variable = new_value

# Using class methods
MyClass.show_class_variable()  # Output: 10

MyClass.change_class_variable(20)
MyClass.show_class_variable()  # Output: 20
```

```python
class MyClass:
    class_variable = "Class Variable"

    def __init__(self, instance_variable):  # INSTANCE METHOD TAKES "SELF" AS A FIRST PARAMETER.
        self.instance_variable = instance_variable

    def instance_method(self):   # NO annotation ( initiated by @: so this method is called on "object"):  # INSTANCE METHOD TAKES "self" AS A FIRST PARAMETER.
        return f"Instance Method: {self.instance_variable}"

    @classmethod  #  with annotation ( initiated by @: so this method is called on  "class" name):  # CLASS METHOD TAKES "cls" AS A FIRST PARAMETER. ---SIMILAR
    def class_method(cls):
        return f"Class Method: {cls.class_variable}"

    @staticmethod   #  with annotation ( initiated by @: so this method is also called on  "class" name): # STATIC METHOD DIRECTLY ACCEPTS THE PARAMETER WITHOUT "cls" AND/OR "SELF" AS A FIRST PARAMETER.--SIMILAR
    def static_method(arg):
        return f"Static Method: {arg}"

# Creating an instance of MyClass
obj1 = MyClass("Instance Var_Value1")


# Accessing instance method
print(obj1.instance_method())  # Output: Instance Method: Instance Variable


# Accessing class method
print(MyClass.class_method())  # Output: Class Method: Class Variable

# Accessing static method
print(MyClass.static_method("Hello"))  # Output: Static Method: Hello
```

# **write an instance method/function to print the first 10 even number**

```python
class Even:
  def __init__(self, num):
    self.num = num
    self.even_result = []
  def even_num(self):
    for i in range(1,self.num):
      if i%2 ==0:
        self.even_result.append(i)
    return self.even_result
```

# **Write an instance method to print the first n prime number:**

```python
class Prime:
    def __init__(self, n):
        self.n = n
        self.result_prime1 = [2]  # Initialize result_prime as an instance variable

    def prime_num(self):
        for i in range(3, self.n + 1):
            for j in range(2, int(i ** 0.5) + 1):
                if i % j == 0:
                    break
            else:
                self.result_prime1.append(i)
        return self.result_prime1
```

```python
obj3 = Prime(20)
obj3.prime_num()
```

# **Write an instance method to print the first n square number:**

```python
class Square:
  def __init__(self,n):
    self.n = n
    self.result_square = []


  def square_num(self):
    for i in range(1,self.n +1):
      sqr_num = i**2
      self.result_square.append(sqr_num)

    return self.result_square
```

```python
obj = Square(11)
obj.square_num()
```

**Write an instance method to print the complete square number less than n:**

```python
class Square:

  def __init__(self,n):
    self.n = n
    self.result_square = []

  def num_square(self):
    for i in range(1,self.n+1):
      if i**2 < self.n:
         self.result_square.append(i)
    return self.result_square
```

```python
obj = Square(10)
obj.num_square()
```

# **Write a program to add digits from a given number:**

```python
def add_digits(n):
  n_str = str(n)

  sum = 0                     #####difference
  for i in range(len(n_str)):
    sum += int(n_str[i])
  return sum
```

# **Given a list, filter all odd elements: index-wise**

```python
def filter_even_odd(list_n):
  even_list = []
  odd_list = []
  for i in range(len(list_n)):
    if list_n[i]%2 ==0:
      even_list.append(list_n[i])
    else:
       odd_list.append(list_n[i])
  return print(f" This is an even list: {even_list}. \n This is an odd list:{odd_list}.")
```

```python
filter_even_odd([1,2,3,4,5,6,7,8,9])
```

# **Given a list, filter all odd elements: element-wise**

```python
def filter_even_odd(list_n):
  even_list = []
  odd_list = []
  for i in list_n:
    if i%2 ==0:
      even_list.append(i)
    else:
       odd_list.append(i)
  return print(f" This is an even list: {even_list}. \n This is an odd list:{odd_list}.")
```
